## Playground to test the Pendulum Environment

```python
import math
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

## Antecedente

```python
angles_list = [math.radians(ele) for ele in np.arange(0, 360, 45)]
angle_position_antecedent = ctrl.Antecedent(np.arange(0, 360, 15), 'angles')
velocity_antecedent = ctrl.Antecedent(np.arange(-8, 8, 1), 'velocity')
```

```python
torque_consequent = ctrl.Consequent(np.arange(-2, 2, 0.1), 'torque')
```

```python
angle_position_antecedent.automf(names=['um', 'dois', 'tres', 'quatro',
                                        'cinco', 'seis', 'sete', 'oito'])
velocity_antecedent.automf(names=['muito_pos', 'pouco_pos', 'quase_zero',
                                  'pouco_neg', 'muito_neg'])
torque_consequent.automf(names=['forte_pos', 'fraco_pos', 'fraquissimo_pos',
                                'fraquissimo_neg', 'fraco_neg', 'forte_neg'])
```

```python
np.arange(0, 360, 45)
```

```python
angle_position_antecedent['um'] = fuzz.trimf(angle_position_antecedent.universe,
                                              [0, 30, 45])
angle_position_antecedent['dois'] = fuzz.trimf(angle_position_antecedent.universe,
                                              [45, 60, 90])
angle_position_antecedent['tres'] = fuzz.trimf(angle_position_antecedent.universe,
                                              [90, 105, 135])
angle_position_antecedent['quatro'] = fuzz.trimf(angle_position_antecedent.universe,
                                              [135, 150, 180])
angle_position_antecedent['cinco'] = fuzz.trimf(angle_position_antecedent.universe,
                                              [180, 195, 225])
angle_position_antecedent['seis'] = fuzz.trimf(angle_position_antecedent.universe,
                                              [225, 240, 270])
angle_position_antecedent['sete'] = fuzz.trimf(angle_position_antecedent.universe,
                                              [270, 285, 315])
angle_position_antecedent['oito'] = fuzz.trimf(angle_position_antecedent.universe,
                                              [315, 330, 360])
angle_position_antecedent['um'].view()
```

```python
velocity_antecedent['muito_neg'] = fuzz.trapmf(velocity_antecedent.universe,
                                               [-8, -8, -6, -5])
velocity_antecedent['pouco_neg'] = fuzz.trimf(velocity_antecedent.universe,
                                               [-6, -4, -1])
velocity_antecedent['quase_zero'] = fuzz.trimf(velocity_antecedent.universe,
                                               [-2, 0, 2])
velocity_antecedent['pouco_pos'] = fuzz.trimf(velocity_antecedent.universe,
                                               [1, 4, 6])
velocity_antecedent['muito_pos'] = fuzz.trapmf(velocity_antecedent.universe,
                                               [5, 6, 8, 8])
velocity_antecedent.view()
```

```python
torque_consequent.automf(names=['forte_pos', 'fraco_pos', 'fraquissimo_pos',
                                'fraquissimo_neg', 'fraco_neg', 'forte_neg'])
torque_consequent['forte_neg'] = fuzz.trapmf(torque_consequent.universe,
                                             [-2, -2, -1.5, -1])
torque_consequent['medio_neg'] = fuzz.trimf(torque_consequent.universe,
                                             [-1.1, -0.80, -0.5])
torque_consequent['fraco_neg'] = fuzz.trimf(torque_consequent.universe,
                                             [-0.55, -0.3, -0.2])
torque_consequent['fraquissimo_neg'] = fuzz.trimf(torque_consequent.universe,
                                             [-0.2, -0.05, 0])
torque_consequent['fraquissimo_pos'] = fuzz.trimf(torque_consequent.universe,
                                             [0, 0.05, 0.2])
torque_consequent['fraco_pos'] = fuzz.trimf(torque_consequent.universe,
                                             [0.2, 0.3, 0.55])
torque_consequent['medio_pos'] = fuzz.trimf(torque_consequent.universe,
                                             [0.5, 0.8, 1.1])
torque_consequent['forte_pos'] = fuzz.trapmf(torque_consequent.universe,
                                             [1, 1.5, 2, 2])
torque_consequent.view()
```
