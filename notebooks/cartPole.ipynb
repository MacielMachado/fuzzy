## Cart Pole

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
theta_antecedent = ctrl.Antecedent(np.arange(-0.420, 0.420, 0.005), 'theta')
omega_antecedent = ctrl.Antecedent(np.arange(-1, 1, 0.001), 'omega')
```

```python
action_consequent = ctrl.Consequent(np.arange(0, 2, 0.001), 'action')
```

```python
theta_antecedent.automf(names=['theta_neg_alto', 'theta_neg_baixo',
                               'theta_pos_baixo', 'theta_pos_alto'])
omega_antecedent.automf(names=['omega_neg', 'omega_pos'])
action_consequent.automf(names=['action_0', 'action_1'])
```

```python
theta_antecedent['theta_neg_alto']=fuzz.trapmf(theta_antecedent.universe,
                                           [-0.42, -0.42, -0.03, -0.02])
theta_antecedent['theta_neg_baixo']=fuzz.trapmf(theta_antecedent.universe,
                                           [-0.03, -0.02, 0, 0])
theta_antecedent['theta_pos_baixo']=fuzz.trapmf(theta_antecedent.universe,
                                           [0, 0, 0.02, 0.03])
theta_antecedent['theta_pos_alto']=fuzz.trapmf(theta_antecedent.universe,
                                           [0.02, 0.03, 0.42, 0.42])
theta_antecedent.view()
```

```python
omega_antecedent['omega_neg']=fuzz.trapmf(omega_antecedent.universe,
                                          [-1, -0.5, 0, 0])
omega_antecedent['omega_pos']=fuzz.trapmf(omega_antecedent.universe,
                                          [0, 0, 0.5, 1])
omega_antecedent.view()
```

```python
action_consequent['action_0']=fuzz.trapmf(action_consequent.universe, [0, 0, 0.5, 1])
action_consequent['action_1']=fuzz.trapmf(action_consequent.universe, [1, 1.5, 2, 2])
action_consequent.view()
```
